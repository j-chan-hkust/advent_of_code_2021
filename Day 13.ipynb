- the column along the fold is "destroyed" and whatever info was there is permanently lost
- there's a transformation that can happen to a point, if its to the left/above a fold, nothing happens, otherwise one of the coords changes

```python
instructions = [instruction.split('=') for instruction in instructions]
```

```python
instructions = [{'dir':instruction[0][-1], 'axis': int(instruction[1])} for instruction in instructions]
```

```python
instructions
```

```python
import itertools

for instruction in instructions:
    for i in range(len(coords)):
        if instruction['dir'] == 'x':
            coords[i][0] = instruction['axis'] - abs(instruction['axis']-coords[i][0])
        elif instruction['dir'] == 'y':
            coords[i][1] = instruction['axis'] - abs(instruction['axis']-coords[i][1])
    coords.sort()
    coords = list(coords for coords,_ in itertools.groupby(coords))
    print(len(coords))
```
