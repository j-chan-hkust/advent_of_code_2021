```python
enhancement_algorithm, image = input_str.split('\n\n')
len(enhancement_algorithm)
```

```python
image = image.split()
```

```python
image = [list(map(lambda char: 0 if char == '.' else 1,row)) for row in image]
```

```python
import numpy as np
image = np.array(image)
image.size
```

```python
def get_binary_code(input_image, x, y):
    binary_num = ''
    for i in [-1,0,1]:
        for j in [-1,0,1]:
            binary_num += str(input_image[i+x][j+y])
    return int(binary_num,2)
```

```python
import copy
def enhance(input_image):
    input_image = np.pad(input_image, 2, 'constant', constant_values=0)
    output_image = copy.deepcopy(input_image)
    for i in range(1,len(input_image)-1):
        for j in range(1,len(input_image[0])-1):
            num = get_binary_code(input_image, i, j)
            if enhancement_algorithm[num] == '#':
                output_image[i][j] = 1
            else:
                output_image[i][j] = 0
    return output_image
```
