```python
input_str = """acedgfb cdfbe gcdfa fbcad dab cefabd cdfgeb eafb cagedb ab | cdfeb fcadb cdfeb cdbaf"""
```

```python
inputs = input_str.split('\n')
```

```python
inputs = [inp.split(' | ') for inp in inputs]
```

```python
count = 0
for _, code in inputs:
    digits = code.split()
    for digit in digits:
        digit_length = len(digit)
        if digit_length == 2 or digit_length == 4 or digit_length == 3 or digit_length == 7:
            count += 1
```

```python
count = 0
for wirings, code in inputs:
    digits = code.split()
    ssds = wirings.split()
    cipher = dict()
    for ssd in ssds:
        if len(ssd) == 2:
            cipher[1] = set(ssd)
        if len(ssd) == 4:
            cipher[4] = set(ssd)
        if len(ssd) == 3:
            cipher[7] = set(ssd)
        if len(ssd) == 7:
            cipher[8] = set(ssd)
    for ssd in ssds:
        if len(ssd) == 6:
            if len(set(ssd).intersection(cipher[1])) != 2:
                cipher[6] = set(ssd)
            elif len(set(ssd).intersection(cipher[4])) == 4:
                cipher[9] = set(ssd)
            else:
                cipher[0] = set(ssd)
    for ssd in ssds:
        if len(ssd) == 5:
            if len(set(ssd).intersection(cipher[1])) == 2:
                cipher[3] = set(ssd)
            elif len(set(ssd).union(cipher[9])) == 7:
                cipher[2] = set(ssd)
            else:
                cipher[5] = set(ssd)
    output = ''
    for digit in digits:
        num = list(cipher.keys())[list(cipher.values()).index(set(digit))] 
        output += str(num)
    count += int(output)
```
