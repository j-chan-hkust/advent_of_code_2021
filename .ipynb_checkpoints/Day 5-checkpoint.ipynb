```python
grid = np.zeros((1000,1000))
```

```python
for line in filter_linear:
    if line['start']['x'] == line['end']['x']:
        # a very convoluted example of grid[0,0:2] = grid[0,0:2] + 1 
        if int(line['start']['y']) < int(line['end']['y']):
            grid[int(line['start']['x']),int(line['start']['y']):int(line['end']['y'])+1] = grid[int(line['start']['x']),int(line['start']['y']):int(line['end']['y'])+1] + 1
        else:
            grid[int(line['end']['x']),int(line['end']['y']):int(line['start']['y'])+1] = grid[int(line['end']['x']),int(line['end']['y']):int(line['start']['y'])+1] + 1
        
    else:
        if int(line['start']['x'])<int(line['end']['x']):
            grid[int(line['start']['x']):int(line['end']['x'])+1,int(line['start']['y'])] = grid[int(line['start']['x']):int(line['end']['x'])+1,int(line['start']['y'])] + 1
        else:
            grid[int(line['end']['x']):int(line['start']['x'])+1,int(line['start']['y'])] = grid[int(line['end']['x']):int(line['start']['x'])+1,int(line['start']['y'])] + 1
```
