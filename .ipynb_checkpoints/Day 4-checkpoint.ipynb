```python
def thats_a_bingo(board):
    for row in board:
        if ''.join(row) == 'xxxxx':
            return True
        
    transpose = [[board[j][i] for j in range(len(board))] for i in range(len(board[0]))]
    
    for col in transpose:
        if ''.join(col) == 'xxxxx':
            return True
            
#     test1 = ''
#     test2 = ''
#     for i in range(len(board)):
#         test1 += board[i][i]
#         test2 += board[i][len(board)-i-1]
#     if test1 == 'xxxxx' or test2 == 'xxxxx':
#         return True
```
