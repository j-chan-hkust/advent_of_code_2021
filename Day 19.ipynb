```
........B..

.........B.
.....S.....
...........

...........

...........
...B.......

....B......
```

As a human, what you do is sort of slide the shapes around, (like they're pieces of transparent paper) until they overlap perfectly in some way.

As code, you can sort of repeat this process by iteratively testing each of the relative positions of S, until you get a perfect match, or there fails to be any overlap of the spaces.

Obviously, the orientation of the areas matters, so what you'd want to test along each of the rotational orientations. In the 2d example, you'd test an area of 2000x2000 different positions for S for a match.

For the 3d examples, the group of orientations is 24 for the 3 cube. There are 2000x2000x2000 possible positions to test

That's 8 billion operations per orientation, not to mention thereare 24 orientations + 33x34/2 permutations of cubes we need to check, not to mention we need to compare the equality of all the points 24ish per cube. Back of the napkin works out to 3 x 10^16 operations.

Even with a clock speed of 3 billion ops per second, this works out to be like 10 million seconds. 

Evidently this isn't going to work, there must be a smarter way.
