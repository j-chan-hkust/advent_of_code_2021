```python
nodes = {}
```

```python
for path in paths:
    start, end = path
    if start in nodes:
        nodes[start].append(end)
    else:
        nodes[start] = [end]
    if end in nodes:
        nodes[end].append(start)
    else:
        nodes[end] = [start]
```

```python
def small(cave):
    if cave == cave.lower():
        return True
    else:
        return False
```

```python
paths = []

candidate_paths = [['start']]

while len(candidate_paths)>0:
    current_path = candidate_paths.pop()
    if current_path[-1] == 'end':
        paths.append(current_path)
        continue
    connected_caves = nodes[current_path[-1]]
    for cave in connected_caves:
        if small(cave) and cave in current_path:
            continue
        else:
            candidate_paths.append(current_path + [cave])
```

```python
def visited_twice(cave, path):
    if path.count(cave)>1:
        return True
    else:
        return False
```

```python
def twice_already(path):
    for node in path:
        if small(node):
            if path.count(node)>1:
                return True
    return False
```

```python
paths = []

candidate_paths = [['start']]

while len(candidate_paths)>0:
    current_path = candidate_paths.pop()
    if current_path[-1] == 'end':
        paths.append(current_path)
        continue
    connected_caves = nodes[current_path[-1]]
    for cave in connected_caves:
        if cave == 'start':
            continue
            
        if not small(cave): # cave big
            candidate_paths.append(current_path + [cave])
            continue
        else: # cave small
            if cave not in current_path:
                candidate_paths.append(current_path + [cave])
            else:
                if twice_already(current_path):
                    continue
                else:
                    candidate_paths.append(current_path + [cave])
            
```
