```python
import copy

def time_step(grid):
    change = False
    new_grid = copy.deepcopy(grid)
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] == '>':
                if j == len(grid[i])-1: # i.e. we are at the right edge of the grid
                    if grid[i][0] == '.':
                        new_grid[i][j] = '.'
                        new_grid[i][0] = '>'
                        change = True
                else: # we aren't at the end
                    if grid[i][j+1] == '.':
                        new_grid[i][j] = '.'
                        new_grid[i][j+1] = '>'
                        change = True
    grid = new_grid
    new_grid = copy.deepcopy(grid)
    
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j] == 'v':
                if i == len(grid)-1: # i.e. we are at the bottom edge of the grid
                    if grid[0][j] == '.':
                        new_grid[i][j] = '.'
                        new_grid[0][j] = 'v'
                        change = True
                else: # we aren't at the end
                    if grid[i+1][j] == '.':
                        new_grid[i][j] = '.'
                        new_grid[i+1][j] = 'v'
                        change = True
    return new_grid, change
```
